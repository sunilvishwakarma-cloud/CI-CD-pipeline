````python
...
# Task 3: Write Unit Tests for an Application

## Objective

Create a simple calculator application, write unit tests for its core functions using Jest, and run those tests automatically in a GitHub Actions CI pipeline.

---

# Step 1: Create the Project

## File: `calculator.js`

```javascript
function add(a, b) {
  return a + b;
}

function subtract(a, b) {
  return a - b;
}

function multiply(a, b) {
  return a * b;
}

function divide(a, b) {
  if (b === 0) {
    throw new Error("Division by zero");
  }
  return a / b;
}

module.exports = {
  add,
  subtract,
  multiply,
  divide
};
```

---

# Step 2: Write Unit Tests

## File: `calculator.test.js`

```javascript
const calculator = require("./calculator");

test("adds 5 + 3 = 8", () => {
  expect(calculator.add(5, 3)).toBe(8);
});

test("subtracts 10 - 4 = 6", () => {
  expect(calculator.subtract(10, 4)).toBe(6);
});

test("multiplies 6 * 7 = 42", () => {
  expect(calculator.multiply(6, 7)).toBe(42);
});

test("divides 20 / 5 = 4", () => {
  expect(calculator.divide(20, 5)).toBe(4);
});

test("throws an error when dividing by zero", () => {
  expect(() => calculator.divide(10, 0)).toThrow("Division by zero");
});
```

---

# Step 3: Configure `package.json`

```json
{
  "name": "calculator-app",
  "version": "1.0.0",
  "scripts": {
    "test": "jest"
  },
  "devDependencies": {
    "jest": "^30.0.0"
  }
}
```

Install dependencies:

```bash
npm install
```

Run the tests locally:

```bash
npm test
```

Expected output:

```
PASS  ./calculator.test.js

Test Suites: 1 passed
Tests:       5 passed
```

---

# Step 4: Integrate with GitHub Actions

Create `.github/workflows/test.yml`:

```yaml
name: Run Unit Tests

on:
  push:
    branches:
      - main
  pull_request:
    branches:
      - main

jobs:
  test:
    runs-on: ubuntu-latest

    steps:
      - uses: actions/checkout@v4

      - uses: actions/setup-node@v4
        with:
          node-version: 20

      - run: npm install

      - run: npm test
```

Whenever code is pushed to the `main` branch or a pull request is opened, GitHub Actions will automatically execute the Jest test suite.

---

# Project Structure

```
calculator-app/
│── calculator.js
│── calculator.test.js
│── package.json
└── .github/
    └── workflows/
        └── test.yml
...
````

````python

# Step 1: Create a Simple Web Application

## File: `app.js`

```javascript
const express = require("express");
const app = express();

app.get("/", (req, res) => {
  res.send("Application is running successfully!");
});

const PORT = 3000;

app.listen(PORT, () => {
  console.log(`Server listening on port ${PORT}`);
});
```

## Install and Run

```bash
npm init -y
npm install express
node app.js
```

The application will be available at:

```
http://localhost:3000
```

---

# Step 2: Create a K6 Performance Test

## File: `load-test.js`

```javascript
import http from "k6/http";
import { check, sleep } from "k6";

export const options = {
  vus: 10,
  duration: "30s",
};

export default function () {
  const res = http.get("http://localhost:3000/");

  check(res, {
    "status is 200": (r) => r.status === 200,
  });

  sleep(1);
}
```

This test simulates:

* 10 virtual users
* Running continuously for 30 seconds
* Sending repeated HTTP GET requests to the application

---

# Step 3: Run the Performance Test Manually

Install K6 according to your operating system.

Run:

```bash
k6 run load-test.js
```

Example output:

```
running (30s), 10 VUs

checks..................: 100.00%
http_req_duration.......: avg=15ms
http_reqs...............: 300
iterations..............: 300
```

---

# Step 4: Automate with Jenkins

Create a Jenkins Pipeline and use the following `Jenkinsfile`:

```groovy
pipeline {
    agent any

    stages {

        stage('Checkout') {
            steps {
                git 'https://github.com/your-username/performance-demo.git'
            }
        }

        stage('Install Dependencies') {
            steps {
                sh 'npm install'
            }
        }

        stage('Build and Start Application') {
            steps {
                sh 'node app.js &'
                sh 'sleep 5'
            }
        }

        stage('Performance Test') {
            steps {
                sh 'k6 run load-test.js'
            }
        }
    }

    post {
        always {
            echo 'Performance testing completed.'
        }
    }
}
```

Whenever Jenkins runs the pipeline:

1. It checks out the source code.
2. Installs dependencies.
3. Starts the application.
4. Executes the K6 load test automatically.

---

# Example CI/CD Flow

```
Developer Pushes Code
          │
          ▼
   Source Code Checkout
          │
          ▼
   Install Dependencies
          │
          ▼
      Build Project
          │
          ▼
   Start Application
          │
          ▼
 Run K6 Performance Test
          │
          ▼
  Report Success/Failure
```

---

# Project Structure

```
performance-demo/
│── app.js
│── package.json
│── load-test.js
└── Jenkinsfile
```
````

````python

# Step 1: Create a Simple Web Application

## File: `app.js`

```javascript
const express = require("express");
const app = express();

app.get("/", (req, res) => {
  res.send("Welcome to the Secure Demo App!");
});

const PORT = 3000;

app.listen(PORT, () => {
  console.log(`Server running on port ${PORT}`);
});
```

## Install Dependencies

```bash
npm init -y
npm install express
```

Run the application:

```bash
node app.js
```

The application will be accessible at:

```
http://localhost:3000
```

---

# Step 2: Install OWASP ZAP

You can install OWASP ZAP from the official website or use the Docker image.

Example using Docker:

```bash
docker run -t owasp/zap2docker-stable zap-baseline.py \
  -t http://host.docker.internal:3000
```

This performs a baseline security scan against the running application.

---

# Step 3: Example OWASP ZAP Output

```
PASS: Application is reachable
PASS: HTTP response received

WARN: Missing security headers
WARN: Information disclosure via headers

FAILURES: 0
```

Review the warnings and update the application if needed.

---

# Step 4: Automate Security Testing with Jenkins

Create a `Jenkinsfile`:

```groovy
pipeline {
    agent any

    stages {

        stage('Checkout') {
            steps {
                git 'https://github.com/your-username/security-demo.git'
            }
        }

        stage('Install Dependencies') {
            steps {
                sh 'npm install'
            }
        }

        stage('Start Application') {
            steps {
                sh 'node app.js &'
                sh 'sleep 5'
            }
        }

        stage('OWASP ZAP Scan') {
            steps {
                sh '''
                docker run --network=host -t owasp/zap2docker-stable \
                zap-baseline.py -t http://localhost:3000
                '''
            }
        }
    }

    post {
        always {
            echo 'Security scan completed.'
        }
    }
}
```

Whenever the pipeline runs, Jenkins:

1. Checks out the source code.
2. Installs dependencies.
3. Starts the web application.
4. Executes an OWASP ZAP baseline scan.

---

# Alternative: Static Code Analysis with SonarQube

If you prefer static analysis, configure SonarQube and add a Jenkins stage such as:

```groovy
stage('SonarQube Analysis') {
    steps {
        sh 'sonar-scanner'
    }
}
```

This analyzes the source code for bugs, code smells, and potential security issues without executing the application.

---

# Example CI Pipeline Flow

```
Developer Pushes Code
          │
          ▼
   Source Code Checkout
          │
          ▼
 Install Dependencies
          │
          ▼
   Build / Start App
          │
          ▼
 Run OWASP ZAP Scan
          │
          ▼
 Publish Results
          │
          ▼
 Build Marked PASS or FAIL
```

---

# Project Structure

```
security-demo/
│── app.js
│── package.json
└── Jenkinsfile
...
````

7. continuous deployment CD

````python

# Step 1: Create a Simple Node.js Application

## File: `app.js`

```javascript
const express = require("express");
const app = express();

app.get("/", (req, res) => {
  res.send("Hello from the Staging Environment!");
});

const PORT = process.env.PORT || 3000;

app.listen(PORT, () => {
  console.log(`Server is running on port ${PORT}`);
});
```

## File: `package.json`

```json
{
  "name": "cd-demo",
  "version": "1.0.0",
  "scripts": {
    "test": "echo Tests passed successfully",
    "start": "node app.js"
  },
  "dependencies": {
    "express": "^4.19.2"
  }
}
```

Install dependencies:

```bash
npm install
```

---

# Step 2: Create a Dockerfile

```dockerfile
FROM node:20-alpine

WORKDIR /app

COPY package*.json ./
RUN npm install

COPY . .

EXPOSE 3000

CMD ["npm", "start"]
```

Build the Docker image:

```bash
docker build -t cd-demo .
```

Run the container:

```bash
docker run -d -p 3000:3000 --name cd-demo-container cd-demo
```

Verify by opening:

```
http://<staging-server-ip>:3000
```

or locally:

```
http://localhost:3000
```

---

# Step 3: Create a Jenkins Pipeline

Create a `Jenkinsfile` in the project root.

```groovy
pipeline {
    agent any

    stages {

        stage('Checkout') {
            steps {
                git 'https://github.com/your-username/cd-demo.git'
            }
        }

        stage('Install Dependencies') {
            steps {
                sh 'npm install'
            }
        }

        stage('Run Tests') {
            steps {
                sh 'npm test'
            }
        }

        stage('Build Docker Image') {
            steps {
                sh 'docker build -t cd-demo .'
            }
        }

        stage('Deploy to Staging') {
            steps {
                sh '''
                docker rm -f cd-demo-container || true
                docker run -d -p 3000:3000 \
                  --name cd-demo-container cd-demo
                '''
            }
        }
    }

    post {
        success {
            echo 'Deployment to staging completed successfully.'
        }

        failure {
            echo 'Pipeline failed.'
        }
    }
}
```

---

# Step 4: CI/CD Pipeline Flow

```
Developer Pushes Code
          │
          ▼
     Source Checkout
          │
          ▼
    Install Dependencies
          │
          ▼
        Run Tests
          │
          ▼
   Build Docker Image
          │
          ▼
 Deploy to Staging Server
          │
          ▼
 Verify Application Access
```

---

# Step 5: Verify the Deployment

After Jenkins finishes successfully:

* Check the running container:

```bash
docker ps
```

* Test the application:

```bash
curl http://localhost:3000
```

Expected response:

```
Hello from the Staging Environment!
```

---

# Project Structure

```
cd-demo/
│── app.js
│── package.json
│── Dockerfile
└── Jenkinsfill
...
````
